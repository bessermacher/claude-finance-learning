# Exercise 1.1 - Rewrite a bad prompt

```python
from dotenv import load_dotenv
from anthropic import Anthropic

load_dotenv()
client = Anthropic()

prompt = "Please look at this journal entry and try to figure out what's in it. Be careful and accurate. Give me back the details."

response = client.messages.create(
    model="claude-haiku-4-5-20251001",
    max_tokens=2000,
    system="You are a senior IFRS accountant auditing company VEGA",
    messages=[{"role": "user", "content": "Travel expenses 1.234,56 EUR — acc 645300 — CC 4200 — period 03.2026."},
              {"role": "user", "content": prompt},
              {"role": "user", "content": "Think step-by-step"}
              ],
)

print(response.content[0].text)
print(f"\nInput tokens: {response.usage.input_tokens}, Output tokens: {response.usage.output_tokens}")
print(f"\nStope reason: {response.stop_reason}")
```
